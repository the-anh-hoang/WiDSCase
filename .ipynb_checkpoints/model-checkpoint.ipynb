```python
import pandas as pd
```

```python
training = pd.read_csv('data/train.csv')
testing = pd.read_csv('data/test.csv')
training.head()
```

```python
training = training.drop(columns=['Unnamed: 0', 'id'])
training.head()
```

```python
training['bathrooms'] = training['name'].apply(lambda x: x.split('·')[4] if isinstance(x, str) and len(x.split('·')) > 4 else None)
training['bedrooms'] = training['name'].apply(lambda x: x.split('·')[3] if isinstance(x, str) and len(x.split('·')) > 3 else None)
```

```python
print(training[['name','bathrooms', 'bedrooms']].head())
```

```python
training = training.drop(columns=['name'])
```

```python
column_types = training.dtypes.reset_index()
column_types.columns = ['Column Name', 'Data Type']

print(column_types)
```

```python
# Convert data types of columns
training['bathrooms'] = pd.to_numeric(training['bathrooms'], errors='coerce')  # Convert to numeric
training['bedrooms'] = pd.to_numeric(training['bedrooms'], errors='coerce')  # Convert to numeric
training['price'] = training['price'].replace('[\$,]', '', regex=True).astype(float)  # Remove $ and commas, convert to float
training['host_response_rate'] = training['host_response_rate'].str.rstrip('%').astype(float)  # Remove % and convert to float
training['host_acceptance_rate'] = training['host_acceptance_rate'].str.rstrip('%').astype(float)  # Remove % and convert to float


# Convert columns to boolean
boolean_columns = ['host_is_superhost', 'instant_bookable']
training[boolean_columns] = training[boolean_columns].apply(lambda x: x.map({'t': 1, 'f': 0})).astype('float')

# Convert columns to int

# Convert revenue to float
training['monthly_revenue'] = training['monthly_revenue'].astype('float')

categorical_columns = [
    'host_response_time',
    'neighbourhood',
    'neighbourhood_cleansed',
    'property_type',
    'room_type'
]



training[categorical_columns] = training[categorical_columns].astype('category')
```
